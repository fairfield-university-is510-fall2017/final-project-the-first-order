### Join All Database Tables

```python
%%sql
SELECT * 
FROM GameStatistics
    JOIN Games ON (Games.GameID=GameStatistics.GameID) 
    JOIN Locations ON (Games.SID=Locations.SID) 
    JOIN Teams ON (Teams.TID=GameStatistics.TeamID) 
    JOIN Quarterbacks ON (Quarterbacks.QBID=GameStatistics.QBID)
```
